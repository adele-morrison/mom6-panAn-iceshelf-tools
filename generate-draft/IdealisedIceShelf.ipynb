# Generate idealised ice thickness file

```python
# import some useful packages
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

```python
x = np.arange(321,800,2) #km
nx = np.arange(240)
y = np.arange(1,4,2) #km
ny = np.arange(2)
thick = np.zeros([2,240])
thick = thick+(500-np.tile(x,(2,1)))
thick = np.where(thick>0,thick,0)
area = np.where(thick>0,(2e3)**2,0)
thick

ds = xr.Dataset(
    data_vars=dict(
        thick=(["ny", "nx"], thick),
        area=(["ny", "nx"], area),
    ),
    attrs=dict(description="idealised triangular ice shelf initalisation."),
)
#ds.to_netcdf('INPUT/testicetrianglefromscratch.nc')
```

```python
ds.thick.plot()
plt.show()
ds.area.plot()
plt.show()
```
